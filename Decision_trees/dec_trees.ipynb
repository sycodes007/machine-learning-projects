```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score, recall_score, matthews_corrcoef, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
df = pd.read_csv('yeast.csv')
df.head()
```

```python
y = df['name']
le = LabelEncoder()
y = le.fit_transform(y)
X = df.drop(columns=['name'])
print('Features shape:', X.shape)
print('Number of classes:', len(le.classes_))
print('Classes:', le.classes_)
```

```python
model = DecisionTreeClassifier(random_state=0)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
y_pred = cross_val_predict(model, X, y, cv=cv)
y_proba = cross_val_predict(model, X, y, cv=cv, method='predict_proba')
```

```python
f1 = f1_score(y, y_pred, average='macro')
recall = recall_score(y, y_pred, average='macro')
mcc = matthews_corrcoef(y, y_pred)
from sklearn.preprocessing import label_binarize
y_bin = label_binarize(y, classes=np.arange(len(le.classes_)))
auc = roc_auc_score(y_bin, y_proba, average='macro', multi_class='ovr')
print('RESULTS')
print('F1 Score (macro):', f1)
print('Recall (macro):', recall)
print('MCC:', mcc)
print('ROC-AUC (macro OVR):', auc)
```

```python
#Bagging(reduce variance=change in prediction/sensitive to training)
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=0,
    n_jobs=-1
)

y_pred_rf = cross_val_predict(rf, X, y, cv=cv)
y_proba_rf = cross_val_predict(rf, X, y, cv=cv, method='predict_proba')

f1_rf = f1_score(y, y_pred_rf, average='macro')
recall_rf = recall_score(y, y_pred_rf, average='macro')
mcc_rf = matthews_corrcoef(y, y_pred_rf)

y_bin = label_binarize(y, classes=np.arange(len(le.classes_)))
auc_rf = roc_auc_score(y_bin, y_proba_rf, average='macro', multi_class='ovr')

print("RANDOM FOREST RESULTS")
print("F1 Score (macro):", f1_rf)
print("Recall (macro):", recall_rf)
print("MCC:", mcc_rf)
print("ROC-AUC (macro OVR):", auc_rf)
```

```python
#Boosting(mistakes)
ada = AdaBoostClassifier(
    n_estimators=200,
    learning_rate=0.05,
    random_state=0
)

y_pred_ada = cross_val_predict(ada, X, y, cv=cv)
y_proba_ada = cross_val_predict(ada, X, y, cv=cv, method='predict_proba')

f1_ada = f1_score(y, y_pred_ada, average='macro')
recall_ada = recall_score(y, y_pred_ada, average='macro')
mcc_ada = matthews_corrcoef(y, y_pred_ada)

auc_ada = roc_auc_score(y_bin, y_proba_ada, average='macro', multi_class='ovr')

print("ADABOOST RESULTS")
print("F1 Score (macro):", f1_ada)
print("Recall (macro):", recall_ada)
print("MCC:", mcc_ada)
print("ROC-AUC (macro OVR):", auc_ada)
```

```python
#Boosting(Error)
gb = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    random_state=0
)

y_pred_gb = cross_val_predict(gb, X, y, cv=cv)
y_proba_gb = cross_val_predict(gb, X, y, cv=cv, method='predict_proba')

f1_gb = f1_score(y, y_pred_gb, average='macro')
recall_gb = recall_score(y, y_pred_gb, average='macro')
mcc_gb = matthews_corrcoef(y, y_pred_gb)

auc_gb = roc_auc_score(y_bin, y_proba_gb, average='macro', multi_class='ovr')

print("GRADIENT BOOSTING RESULTS")
print("F1 Score (macro):", f1_gb)
print("Recall (macro):", recall_gb)
print("MCC:", mcc_gb)
print("ROC-AUC (macro OVR):", auc_gb)
```
